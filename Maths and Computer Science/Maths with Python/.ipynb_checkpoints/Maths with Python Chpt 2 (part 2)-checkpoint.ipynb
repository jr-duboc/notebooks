```python
'''
Plotting gravitional force between two objects (mass 0.5kg and 1.5kg) 
relative to the distance between them
'''

import matplotlib.pyplot as plt

# Draw graph
def draw_graph(x,y):
    plt.plot(x,y,marker='o')
    plt.xlabel('Distance in metres')
    plt.ylabel('Gravitional force in newtons')
    plt.title('Gravitional force relative to distance')
    plt.show()
    
def generate_F_r():
    # Generate values for range r
    r = range(100,1001,50)
    # Store generated values for F in a list
    F = []
    
    # Gravitional Constant G
    G = 6.674*(10**-11)
    
    # Two masses
    m1 = 0.5
    m2 = 1.5
    
    # Calculate force
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force)
        
    # Generate the graph
    draw_graph(r,F)
    
generate_F_r()
```

```python
'''
Projectile motion
'''
import matplotlib.pyplot as plt
import math

# Draw the graph
def draw_graph(x,y):
    plt.plot(x,y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('Projectile motion of a ball')

# Generate floating point range
def frange(start,final,increment):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + increment
        
    return numbers

# Trajectort (u is velocity, theta (θ) is the angle)
def draw_trajectory(u, theta):
    theta = math.radians(theta)
    g = 9.8
    
    # Time of flight (2 times peak)
    t_flight = 2*u*math.sin(theta)/g #twice the peak time u x sin(θ) / g
    # Define time intervals
    intervals = frange(0, t_flight,0.001)
    
    # List of x and y coordinates
    x = []
    y = []
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
        
    draw_graph(x,y)
    
# Comparing different velocities
u_list = [20,30,40]
theta1 = 45
for uItem in u_list:
    draw_trajectory(uItem,theta1)

plt.legend(['20', '40', '60'])    
plt.show()

try:
    u = float(input('Enter the intial velocity in m/s:' ))
    theta = float(input('Enter the angle of projection in degrees: '))
except ValueError:
    print('Invalid Input')
else:
    draw_trajectory(u, theta)
    plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(-200,201)
y_values = []
for x in x_values:
    # Calculate the value of the quadratic function
    y = x**2 + 2*x + 1
    y_values.append(y)
    
plt.plot(x_values, y_values)
plt.title('Quadratic equation: y=x^2 + 2x + 1')
plt.show()
```
