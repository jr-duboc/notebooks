```python
import matplotlib.pyplot as plt
import matplotlib.dates as datesNum
from datetime import date

waterReadings = [
    [date(2017,9,21),  20.05],
    [date(2017,10,5),  25.10],
    [date(2017,10,26), 31.30],
    [date(2017,11,10), 37.00],
    [date(2017,12,1),  45.50],
    [date(2017,12,28), 56.30]
]

dates = []
readings = []
for reading in waterReadings:
    dates.append(reading[0])
    readings.append(reading[1])

plt.title('Water readings')
plt.xticks(rotation=-60)
plt.plot(dates,readings)
plt.show()

'''
plt.title('Electricity readings')
plt.xticks(rotation=-60)
plt.plot(dates,elecReadings)
plt.show()
'''
#plt.rcParams['figure.figsize'] = [20,20]

```
